# Código para geração de caso do Zumbi Health a partir de um CSV

```python
import pandas as pd
import numpy as np
```

```python
def CSVtoCase(filename):
    template = '''# {numero} (case)

![zumbi_{numero}](https://www.ic.unicamp.br/~santanch/lab/zombie-health/images/gen02/zumbi_{numero}.png)

{nome} se apresentou na clínica reclamando que algum bicho o picou ou mordeu, o que você deseja examinar?

* Local da Picada -> Mostra Sintoma({dor} {sangramento} {inchaco})
* Estado do Zumbi -> Mostra Sintoma({estado})
* Ferida -> Mostra Sintoma({necrose})
* Batimentos -> Mostra Sintoma({arritmia})

* Realizar Diagnóstico -> Diagnóstico {diagnostico}
'''
    
    patient = ""
    
    df = pd.read_csv(filename)
    
    cases = df.values
    
    for c, value in enumerate(cases):
        
        dor = "sente muita dor no local da mordida" if (value[1] == 1) else "sente pouca dor no local da mordida"
        sangramento = "; o local está mudando para uma coloração azul" if (value[2] == 1) else ""
        inchaco = ""
        if (value[3] == 1):
            if (value[9] == "jararaca"):
                inchaco = "; local muito inchado"
            elif (value[9] == "aranha marrom"):
                inchaco = "; local inchado"
            else:
                inchaco = "; local levemente inchado"
                
        estado = "está com o olho caído" if (value[4] == 1) else ""
        if (value[4] == 1 and value[5] == 1):
            estado += "; "
        estado += "tem dificuldade de engolir" if (value[5] == 1) else ""
        if (len(estado) > 0 and value[6] == 1):
            estado += "; "
        estado += "está cansado para respirar" if (value[6] == 1) else ""
        if (estado == ""):
            estado = "está com os olhos normais; respira e engole de forma regular"
        
        necrose = "tem uma ferida grande, com bordas elevadas e tem uma área preta no meio" \
                     if (value[7] == 1) else "tem uma ferida sem destaques além daqueles já descritos"
            
        arritmia = "tem batimentos cardíacos descompassados, num ritmo atrapalhado" \
                     if (value[8] == 1) else "tem batimentos cardíacos normais"
        
        print(template.format(numero=c+1, nome=value[0], dor=dor, sangramento=sangramento, inchaco=inchaco,
                              estado=estado, necrose=necrose, arritmia=arritmia, diagnostico=value[9]))
        
    return cases
```
